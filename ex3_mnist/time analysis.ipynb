```python
import numpy as np
import matplotlib.pyplot as plt
import sys;sys.path.append('../utils/')
import imp
import plot_tools,api
import pandas as pd
import networkx as nx
from measures import matchY,measures_calculator,matchY
import time
```

```python
N=[1e3,5e3,1e4,2e4,3e4,4e4,5e4,6e4,7e4,8e4,9e4,1e5]
time1=[]
time2=[]
time3=[]
time4=[]
for n in N:
    X=np.random.rand(int(n),3)
    Y_pred,t1,t2,t3,t4,t5=api.DGSFC.fit(  X,
                  K_d=50,
                  K_s=25,
                  alpha=0.9,
                  epsilon=0,
                  plot=False,
                  scale=True
                 )

    time1.append(t2-t1)
    time2.append(t3-t2)
    time3.append(t4-t3)
    time4.append(t5-t4)
```

```python
time_change_n=np.array([time1,time2,time3,time4])
np.save('./time_change_n.npy',time_change_n)
```

```python
plt.figure(figsize=(7,5),dpi=600)
plt.plot(N,time1,marker='o',c='C1',label='density estimation')
plt.plot(N,time2,marker='o',c='C2',label='local clusters')
plt.plot(N,time3,marker='o',c='C3',label='topo-graph')
# plt.plot(N,time4,marker='o',c='C4',label='extract clusters')
# plt.xlabel('n',fontsize=15)
# plt.ylabel('time(seconds)',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=15)
plt.savefig('time_change_n.png',bbox_inches = 'tight')
```
